# Can you calculate $\pi$?
The python function `pi()` samples uniform distributions for two sets of random numbers $0 \leq x \leq 1$ and $0 \leq y \leq 1$. The total number of samples is `sample_size`. The code also reports the number of data points, `count_circle`, for which $x^2 + y^2 \leq 1$ and those data poits are stored in the dictionary `xy`. Below is a graph of the sample points in the $(x,y)$ plane.

<img src="/work/pi_graph.png" width="400"/>

The purple points are those for which $x^2 + y^2 \leq 1$ holds true. The points in green are those for which the $x^2 + y^2 \leq 1$ condition is **not** met. These points are stored in the dictionary `xy_out`.

```
def pi(sample_size):

    x = np.random.random(sample_size)
    y = np.random.random(sample_size)
    xy     = {'x':[], 'y':[]}
    xy_out = {'x':[], 'y':[]}

    count_circle = 0

    for i in range(sample_size):
        r = x[i]**2 + y[i]**2

        if r <= 1:
            xy['x'].append(x[i])
            xy['y'].append(y[i])
        else:
            xy_out['x'].append(x[i])
            xy_out['y'].append(y[i])
    
    return(xy, xy_out)
```
